```python
import numpy as np
import pandas as pd

data = pd.read_csv('...\\Predicting Pulsars\\pulsar_stars.csv')
```

Data yang digunakan untuk dianalisis disini adalah data yang berasal dari kaggle dengan link berikut https://www.kaggle.com/pavanraj159/predicting-a-pulsar-star. Data ini adalah data-data pengamatan pulsar, atau #neutron star yang berotasi dengan sangat cepat. Langkah pertama yang tentunya harus dilakukan adalah meng-import atau memasukkan data-data tersebut ke dalam kernel kita.   

Setelah data-data tersebut sudah berhasil di masukkan ke dalam kernel, tentu kita juga perlu mengetahui bentuk dari data-data tersebut. Disini saya menggunakan fungsi #head yang bisa menunjukkan 10 data teratas beserta dengan masing-masing nama kolomnya. 

```python
data.head(10)
```

Setelah dilihat bentuk data-data ini, bisa dilihat bahwa nama-nama kolomnya masih terlalu sulit dan panjang. Untuk memudahkan analisis nantinya, nama-nama tersebut dirasa perlu diganti dengan nama-nama yang lebih jelas dan lebih pendek.

```python
data.columns = ['mean', 'standard_deviation', 'excess_kurtosis',
                'skewness', 'mean_DM', 'standard_deviation_DM',
                'excess_kurtosis_DM', 'skewness_DM', 'target']
data.head(5)
```

Kemudian data-data tersebut perlu di ubah ke dalam skala 0 sampai 1. Hal ini dilakukan untuk memudahkan analisis menggunakan #machine learning nantinya

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
data_2 = data

x = np.reshape(np.array(data['mean']), [-1, 1])
scaler = MinMaxScaler().fit(x)
data_2['mean'] = scaler.transform(x)
```

```python
col = data.columns

for x in range(8):
    y = np.reshape(np.array(data[col[x]]), [-1, 1])
    scaler = MinMaxScaler().fit(y)
    data_2[col[x]] = scaler.transform(y)
```

```python
data_2.head(5)
```

Setelah itu, data-data tersebut perlu dibagi menjadi dua berdasarkan targetnya. Jika target menunjukkan angka 0 itu artinya objek yang diamati tersebut bukan #neutron star, sedangkan jika bernilai 1 maka objek tersebut adalah #neutron star.

```python
stars = data_2[data_2['target'] == 1]
nonstars = data_2[data_2['target'] == 0]
```

```python
stars_target = stars.target
nonstars_target = nonstars.target

stars = stars.drop(['target'], axis=1)
nonstars = nonstars.drop(['target'], axis=1)
```

Lalu, langkah berikutnya yang saya ambil adalah membagi data menjadi dua kelompok, kelompok train dan kelompok test. Data yang berada pada kelompok train adalah data yang akan digunakan untuk melatih sistem #machine learning sedangkan data pada kelompok test adalah data yang akan menguji sistem #machine learning tersebut.

```python
from sklearn.model_selection import train_test_split

x1_train, x1_test, y1_train, y1_test = train_test_split(stars, stars_target, test_size=0.3)
x2_train, x2_test, y2_train, y2_test = train_test_split(nonstars, nonstars_target, test_size=0.3)
```

```python
x_train = pd.concat([x1_train, x2_train])
x_test = pd.concat([x1_test, x2_test])
y_train = pd.concat([y1_train, y2_train])
y_test = pd.concat([y1_test, y2_test])
```

Kemudian, saya memasukkan semua #library machine learning yang saya butuhkan. 

```python
from sklearn.linear_model import LogisticRegression, SGDClassifier, LinearRegression
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, ExtraTreesClassifier, VotingClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
Analisis pun dimulai.
```

```python
test = [LogisticRegression(), KNeighborsClassifier(), SVC(kernel='linear', C=0.025), SGDClassifier(max_iter=1000),
        LinearRegression(),DecisionTreeClassifier(), RandomForestClassifier(), MLPClassifier(max_iter=1000),
        AdaBoostClassifier(), GaussianNB()]
score = np.zeros([len(test), 1])
```

```python
for clf, x in zip(test, range(len(test))):
    classifier = clf.fit(x_train, y_train)
    score[x] = classifier.score(x_test, y_test)
```

```python
score
```

```python
names = ['Logistic Regression', 'KNN', 'SVM', 'SGD Classifier', 'Linear Regression',
         'Decision Tree', 'Random Forest', 'Neural Network', 'AdaBoost Classifier',
         'Naive Bayes']
names = pd.DataFrame(np.transpose(np.array(names)))
```

```python
score = pd.DataFrame(score)
scores = pd.concat([names, score], axis=1)
scores.columns = ['Methods', 'Score']
```
